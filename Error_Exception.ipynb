# What is an Exception:

- An exception is an unwanted events.
- It is an that occurs during program execution (runtime error).
- Examples : 
    * ZeroDivisionerror  - it occurs when we divide number from zero
    * keyerror - it occurs when particular index is not present
    * AttributeError   - it occurs when invalid object method
    * ValueError  - invalid value
    * TypeError  - wrong data type
    * NameError  - variable not defined

gaierror - Internet not found

## How to handle these errors :

These error can be handle by using try and except block 

- try : It is used to find the error.
- except : It handles the error
- finally : It executes always, whether exception occur or not
- If we use try and except then the normal flow of program will not break
### synatx :

try :
    statement
except :
    statement

```python
a = 10
print(a/0)
print("Suhani")
print("Anukul")
```

```python
a =10
try : 
    print(a/0)
except ZeroDivisionError:
    print("Number can't be divided by zero")
    

print("Suhani")
print(7+8)
```

```python
a =10
try : 
    print(a/0)
except ZeroDivisionError:
    print("Number can't be divided by zero")
finally:
    print("Program completed")

# print("Suhani")
# print(7+8)
```

```python
try : 
    print(10/0)
except Exception as e:
    print(e)
```
