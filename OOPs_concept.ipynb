# Inheritance 

- Inheritance allows one class to acquire properties and methods..
- in this first class is known as parent class, super class...
- Second class is known as child class, sub class...
- there are 5 type of inheritance
     * Single level inheritance
     * Multilevel inheritance
     * Multiple inheritance
     * hierarachical inheritance
     * Hybrid inheritance

1. Single level Inheritance : In this one child class inherits the properties and behaviour from its parent class..





```python
class Animal:
    def eat(self):
        print("Animal Eats")

class Dog: 

    def bark(self):
        print("Dog barks")

obj = Dog()
obj2 = Animal()
obj2.eat()
obj.bark()
```

```python
class Animal: # this is a parent class
    def eat(self):
        print("Animal Eats")

class Dog(Animal): # here we are inheriting ANimal class in dog class. This is a child class

    def bark(self):
        print("Dog barks")

obj = Dog()
obj.eat()
obj.bark()
```

```python
# Multilevel Inheritance : A child class inherits properties from another child class
"""
class A
|
class B(A)
|
class C(B)
"""


class A:
    def m1(self):
        print("Hi from m1")
class B(A):
    def m2(self):
        print("Hi from m2")
class C(B):
    def m3(self):
        print("Hi form m3")


obj = C()
obj.m1()
obj.m2()
obj.m3()
```

```python
# Multiple Inheritance :It means one child class inherits from multiple parents.

""" 
class A      class B
    \          /
       class C
"""


class Father:
    def money(self):
        print("Father gives Money")
class Mother:
    def love(self):
        print("Mother gives love")
class Suhani(Father,Mother):
    def study(self):
        print("Suhani studies")

obj = Suhani()
obj.money()
obj.love()
obj.study()
```

```python
# Hierarchical Inheritance : One parent class has multiple child classes..

"""
       Parent
      /       \
    Suhani    Brother/Sister
"""

class Parent:
    def money(self):
        print("Gives Money")
class Suhani(Parent):
    def study(self):
        print("studies")

class Brother(Parent):
    def task(self):
        print("timepass")

obj_1 = Suhani()
obj_2 = Brother()
obj_1.study()
obj_2.task() 
```

```python
# Hybrid Inheritance : It is a combination of two or more inheritance types..

class A:
    pass
class B(A):
    pass

class C(A):
    pass

class D(B,C):
    pass

obj = D()
```

# Encapsulation

- Encapsulation is a process of combining data and methods into a single unit(class) and provide controlling access to the data.

## Access modifiers in Python :

    * Public - name : It can be accessible everywhere
    * Protected - _name : Used in intended class or subclass, (created by using single underscore)
    * Private - __name : used in specific class, (created by using double underscore)

```python
# Public modifiers:

class demo:
    def __init__(self):
        self.name = "Suhani"

s= demo()
print(s.name)
```

```python
# Protected 

class demo:
    def __init__(self):
        self._marks = 90

s= demo()
print(s._marks) #Cannot update value
```

```python
# Private :

class demo:
    def __init__(self):
        self.__marks = 90

s= demo()
print(s.__marks) 
```

```python
# Accessing Private variable : By using a public method (getter)

class demo:
    def __init__(self):
        self.__marks = 90
    def mr(self):
        return self.__marks

obj = demo()
obj.mr()
```

# Abstraction

- It is a process of hiding the implementation detail and display only necessary functionality to user.
- The user knows what an object does, but does not need to know how it does it..
- In python it is implemented by using abc module where abc stands for Abstract Base Class.
- Abstract class acts as a blueprint.
- Abstract should not have any implementation detail
- We can't create object of abstract class

```python
# create a abstract method

from abc import ABC,abstractmethod

@abstractmethod
def demo(self):
    pass
```

```python
# creating first abstract function

from abc import ABC,abstractmethod

class shape(ABC):

    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def stop(self):
        pass

class Circle(shape):
    def area(self):
        print("Area of circle")

obj = Circle()
obj.area()
```

# Polymorphism 

- It is one of the four pillar of OOPs
- The word polymorphsim comes from two Greek words
- Poly = Many & Morph = Forms
- Polymorphism means One interface with many forms.

## Types of Polymorphism  

1. Compile time Polymorphism 
2. Runtime Polymorphism

- Python doesn't support traditional compile-time method overloading line in Java, C++..

### How Polymorphsim can be achieved :

- Method Overriding(Runtime Polymorphism)
- Operator Overloading.
- Function Polymorphism (same function works with different data types)

```python
class Demo:
    def s1(self):
        print("Hi from Suhani")
class Demo2(Demo):
    def s1(self):
        print("Hi from Anukul")
class Demo3(Demo):
    def s1(self):
        print("Hi from sweet 16")

obj = Demo2()
obj2 = Demo3()

obj.s1()
obj2.s1()
```

```python
# example Loops
class Demo:
    def s1(self):
        print("Hi from Suhani")
class Demo2(Demo):
    def s1(self):
        print("Hi from Anukul")
class Demo3(Demo):
    def s1(self):
        print("Hi from sweet 16")


obj = [Demo(),Demo2(),Demo3()]
for i in obj:
    i.s1()
```

```python
# Method Overloading :

class Demo:

    def add(self,a,b):
        print(a+b)
    def add(self,a,b,c):
        print(a+b+c)


obj = Demo()
obj.add(1,2)
```

```python
# Method Overloading :

class Demo:

    def add(self,a,b):
        print(a+b)
    def add(self,a,b,c=0): #Here C=0 assigned as default value.
        print(a+b+c)


obj = Demo()
obj.add(1,2)
```
