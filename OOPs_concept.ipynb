# OOPs..

- It stands for Object oriented Programming system.
- It is approach in which data and functions are combined in a single unit called as object
- An object represent real-world entity : 
    * Data(Attribute/properties)
    * function (Methods/behaviour)

- This we use to maintain, organized and reuable code..
- There are 4 pillars of OOPS :
    * Inheritance
    * Encapsulation
    * Polymorphism
    * Abstraction

## Import Concepts : 

- class  - Blueprint or a template
- object - instance of class
- Atrributes - variable inside a class
- Method - Functions inside a class
- Constructor - Special methods called automatically
- Inheritance - Acquiring properties from another class
- Polymorphism - One interface, many forms
- Encapsulation - Wrapping data and methods together
- Abstraction - hiding implementation detail
