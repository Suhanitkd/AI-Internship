# *args and **kwargs 

- *args : Accepts multiple position arguments.
- **kwargs : Accepts multiple keyword argument

```python
def add(a,b):
    print(a+b)

add(10,10,10)
```

```python
def demo(*args):

    for i in args:
        print(i)


demo(1,2,3,4,5,6,7,0,8,9)
```

```python
def add(*args):
    sum = 0
    for i in args:
        sum +=i
    print(sum)


add(10,20)
add(10,20,30)


```

```python
def test(**kwargs):
    print(kwargs)


test(name="Suhani",age=22,city="Varanasi")
```

## program for counting character without using any function

```python
st = str(input("Enter the string:"))
ch_count = []
# ch = ""
for ch in st:
    
        if ch in ch_count:
            ch_count +=1
        else :
            ch_count =1
print("Occurance of character.=")
for i in ch_count:
    print(i,"=",ch_count[i])
```

# Find the largest element in a list without using any pre-defined function.

```python
l =[10,45,20,78,12]
largest = l[0]
for i in l:
    if i > largest:
        largest = i
print ("the largest number is:" ,largest)
```

# Remove Duplicates from list without using any pre-defined function

```python
li = [1,1,5,8,9,8,6,5,3,7]
num = []
for i in li:
    if i not in num:
        num.append(i)
print (num)
```

# Reverse a list without using built-in function..

```python
li= [10,20,30,40]
n = []
for i in range(len(li)-1,-1,-1):
    n.append(li[i])

print(n)
```

# saperating +ve and -ve numbers from a list

```python
num = [-1,-5,-6,5,6,8,9,-7,6,-2,7]
a =[]
b =[]
for i in num:
    if i < 0:
        a.append(i)
    elif i > 0:
        b.append(i)

print(a,b)
```

# Check whether a tuple is a palindrome

```python
tu = (10,20,30,20,10)
if tu == tu[::-1]:
    print ("the tuple is palindrome")
else:
    print ("the tuple is not palindrome")
```
