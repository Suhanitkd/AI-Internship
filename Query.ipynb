# *args and **kwargs 

- *args : Accepts multiple position arguments.
- **kwargs : Accepts multiple keyword argument

```python
def add(a,b):
    print(a+b)

add(10,10,10)
```

```python
def demo(*args):

    for i in args:
        print(i)


demo(1,2,3,4,5,6,7,0,8,9)
```

```python
def add(*args):
    sum = 0
    for i in args:
        sum +=i
    print(sum)


add(10,20)
add(10,20,30)


```

```python
def test(**kwargs):
    print(kwargs)


test(name="Suhani",age=22,city="Varanasi")
```

# Check whether a tuple is a palindrome

```python
tu = (10,20,30,20,10)
if tu == tu[::-1]:
    print ("the tuple is palindrome")
else:
    print ("the tuple is not palindrome")
```

```python
sentence = input("Enter a sentence: ")

words = sentence.split()   # Split sentence into words
longest_word = ""          # Initialize longest word as empty

for word in words:
    if len(word) > len(longest_word):
        longest_word = word

print("The longest word is:", longest_word)
print("Length of longest word:", len(longest_word))
```
