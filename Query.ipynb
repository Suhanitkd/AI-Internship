# *args and **kwargs 

- *args : Accepts multiple position arguments.
- **kwargs : Accepts multiple keyword argument

```python
def add(a,b):
    print(a+b)

add(10,10,10)
```

```python
def demo(*args):

    for i in args:
        print(i)


demo(1,2,3,4,5,6,7,0,8,9)
```

```python
def add(*args):
    sum = 0
    for i in args:
        sum +=i
    print(sum)


add(10,20)
add(10,20,30)


```

```python
def test(**kwargs):
    print(kwargs)


test(name="Suhani",age=22,city="Varanasi")
```

## Write a program to find the longest word in a sentence.

```python
sentence = input("Enter your sentence:")
words = sentence.split()
# print(words)
lamba_word=''
for i in words:
    if len(i) > len(lamba_word):
        lamba_word = i

print("Longest Word in the Sentence is", lamba_word)
print("Length of the Word is ", len(lamba_word))
```

## Write a program to reverse each word in a sentence.

```python
sentence = input("Enter the sentece: ")
words = sentence.split()
reverse = []
for i in words:
    reverse.append(i[::-1])
result = " ".join(reverse)
print("Sentence with each word reversed ",result)
```
