# Comments : 

- Comments are lines of text in python program that are ignored by python interpreter.
- This we use to explain the code

- Comments are non-executable statments used to describe or explain the code..

## Types of Comments : 

In python there is only single-line comment.. which is applied by using # symbol..
But we can add multi line comment also by using """""" or '''''' (It is denoted as multiline string which we use to document our code)
