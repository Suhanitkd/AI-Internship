    if condition :

```python
num = 4
if  num>5:
    print("Yes number is greater than 5")
```

# Take a character from user and check whether it is a vowel or consonant

```python
ch = input("Enter the alphabet to be checked:")
if ch =='a' or ch=='i' or ch=='e' or ch=='o' or ch=='u' or ch=='A' or ch=='E' or ch=='I' or ch=='O' or ch=='U':
    print(f'{ch} is vowel.')
else :
    print(f'{ch} is consonant')
```

# elif :

- It is a keyword in python which which to check condition and it works only when we have used if block. 
- In python there can be multiple elif, but there should a single if block

#### syntax : 

    if condition :
        statement
    elif condition :
        statement
    elif condition :
        statement

```python
num = int(input("Enter a number"))
if num > 0:
    print("Number is +ve")
elif num < 0:
    print("numebr is negative")
elif num == 0:
    print("Number is zero")
```

```python
a = 67
b = 82
c = 12
if a>b and a>c :
    print(f'{a} is the greatest number.')
elif b>a and b>c :
    print(f'{b} is the greatest number.')
else :
    print(f'{c} is the greatest number.')
```
