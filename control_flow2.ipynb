# range() : 
- It is used to print numbers from any specified sequence. In this there are 3 attributes : start,stop, step
- Step size is optional,
- whatever stop point we give it prints till stop-1. Example : stop =10 it will print till 9 only
- Intially range function with positive index.

```python
for (int i =1;i<11;i++)
{
    printf("%d",i)
}
```

```python
for i in range(1,11):
    print(i)
```

```python
for i in range(10,0,-1):
    print(i)
```

```python
for i in range(2,21,2):
    print(i)
```

```python
for i in range(1,51,2):
    print(i)
```

```python
for i in range(49,0,-2):
    print(i)
```

# While Loop : It is also used to repeat a block of code, but executes until the condition otherwise it get stop.

    syntax of while loop :
        initialization
        while condition:
            statement
            increment
            

```python
i = 1
while i<11:
    print(i)
    i +=1
```

```python
l=[1,2,3,4,5,6,7,8,10,9,22,13,15,65,43]
i = 0
while i<15:
    if l[i] %2 ==0: 
        print(l[i])
    i +=1
```

# WAP to print all vowels from inputted string

```python
st = input("Enter a string")
v = "aeiouAEIOU"
i=0
while i<len(st):
    if st[i] in v :
        print(st[i])
    i +=1
```

```python
st = input("Enter a string")
v = "aeiouAEIOU"
for i in st:
    if i in v:
        print(i)
```
