# WAP for "*" based question

```python
for i in range (1,6): # rows
    for j in range (0,i): # columns
        print("*",end=" ")
    print("")
```

```python
for i in range (5,0,-1):
    for j in range (0,i):
        print("*",end=" ")
    print("")
```

```python
for i in range (5,0,-1):
    for j in range (0,i):
        print(" ",end=" ")
    for j in range (i):
        print("*",end=" ")
    print("")
```

```python
n = 5
for i in range(n):
    for j in range(i,n):
        print(' ',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()   
```

```python
rows = int(input("Enter a number of rows")) #...??(iteration) 5

for i in range(rows): # this we use to access rows
    for j in range(rows-i-1): # This loop is for printing leading spaces
        print(" ",end=" ")
    for k in range(2*i+1): # This loop is used for printing the stars..
        print("*",end =" ")
    print()
```

# Sum of digit

```python
num = int(input("Enter a Number:"))
print("original number", num)
sum=0
while num>0:
    a = num % 10
    sum += a
    num = num//10

print(sum)
    
```

# Reverse a Number

```python
num = int(input("Enter a Number:"))
print(num)
rev=0
while num>0:
    a = num%10
    rev = rev*10 +a
    num //=10

print(rev)
```

# Wap to check whether number is palindrome or not..

```python
num = int(input("Enter the number:"))
print(num)
num2 = num
pal=0
while num>0:
    a = num%10
    pal = pal*10 +a
    num //=10

if pal == num2 :
    print ("The number is palindrome")
else :
    print ("The number is not palindrome")
print(pal)
```

## break and continue

- break statment : It is used to terminate the loop.
- continue statment : It is used to skip the particular process

```python
for i in range(1,10):
    if i ==5:
        break
    print(i)
```

```python
for i in range(1,10):
    if i == 6:
        continue
    print(i)
```
