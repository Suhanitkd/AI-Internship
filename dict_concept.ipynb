# Dictionary In Python:

- It is a built-in data type in python used to store data in key-value pair
- It is hetrogenous
- Ordered collection of item 
- created by using curly braces{}
- key must be unique
- values can be duplicated
- it is also mutable
- No indexing(data can be accessed only by key).

# Traversing

```python
# To print keys
sq = {1:1,2:4,3:9,4:16,5:25}
for i in sq:
    print(i)
```

```python
# To print values
for i in sq.values():
    print(i)
```

```python
for i,j in sq.items():
    print(i,j)
```
