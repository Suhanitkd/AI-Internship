# File Handling 

- File handling is a process of creating,opening,reading, writing, updating,deleting the files using python..
- it allows data to be stored permanently on the system..

## Why do we need file handling :

- Without file handling, data is stored in a variable and lost when the program ends

### Types of files : 

1. Text file : These files store data in character format. (Exmaple : .txt, .csv, .json, .xml, .html, .py, .ipynb etc)
2. Binary File : It stores data in binary format (0 and 1) (Example : .jpg, .png, .mp3, .mp4, .pdf, .exe, etc.)

### File Operations
 * Create
 * Open
 * Read
 * Write
 * Append
 * Update
 * Close
 * Delete



# File Modes:

- r  : Read only
- w  : Write(creates file if it doesnt exist, overwrite existing data)
- a  : Append (adds data at the end)
- x  : Create a new file(fails if file exists)
- r+ : Read and Write
- w+ : Read and write(overwrites existing data)
- a+ : Read and Append
- rb : Read binary
- wb : Write binary
- ab : Append binary

```python
# Writing into a file : 
file = open("new_file.txt","w")
file.write("Hii from Anukul Sir")
file.close()
```

```python
# Writing Multiple Lines
file = open("random.txt","w")
file.write("Anukul\n")
file.write("20\n")
file.write("Varanasi\n")
file.close()
```

```python
file = open("random.txt","w")
data= ["Sarika Srivastava\n",
       "Sweet 16\n",
       "Varanasi"
       ]
file.writelines(data)
file.close()
```
