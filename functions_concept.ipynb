```python
def prime(a):
    count = 0
    for i in range(2,a+1):
        if a%i==0:
            count +=1
    if count==1:
        print(f'{a} the number is prime.')
    else :
        print("the number is not prime.")


prime(96)
    
```

# Calculator program using function:

1. for addition
2. for substraction
3. multiplication
4. for division

4+5+7+8-2*6

```python

def calculator(ch,a,b):
    if ch==1:
        print(a+b)
    elif ch==2:
        print(a-b)
    elif ch==3:
        print(a*b)
    elif ch==4:
        print(a/b)
    else:
        print("Number not entered from the choice given.")

while True:

    
    option = '''
    1. Addition
    2. Substraction
    3. multiplication
    4. Division
    '''
    print(option,"\n","from these option select one")
    ch = int(input("Enter a choice"))
    num = int(input("ENter a number"))
    num2 = int(input("ENter another number"))

    calculator(ch,num,num2)

    p = input("Do you want to continue? Y or N")
    if p=='N' or p== 'n':
        print("The process is complete.")
        break
    
```

```python
def Calculator(result,a,ch):
    if ch=='+':
        return result+a
    elif ch=="-":
        return result-a
    elif ch=='*':
        return result*a
    elif ch=='/':
        return result/a
    else :
        print ("Invalid choice.")

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
ch = input("Enter your choice from +,-,*,/")
result = Calculator(a,b,ch)
print(result)
while True:
    choice = input("Either you want to continue? Y or N")
    if choice == 'n' or choice=='N':
        print ("Calculator close.")
        break
    ch = input("Enter your choice from +,-,*,/")
    b = int(input("Enter the second number: "))
    result = Calculator(result,b,ch)
    print(result)
```

## Program for Lcm

```python
def lcm(a,b):
    if (a>b):
        greater = a
    else :
        greater = b

    while True:
        if (greater %a==0) and (greater %b==0):
            result= greater
            break
        greater +=1
    return result
num1 = input("Enter the first number:")
num2 = input("Eneter the second number:")
lcm(num1,num2)
print()
```
