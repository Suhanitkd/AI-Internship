# List:

- A list is one of the most commonly used data type/ data str. in Python
- It is used to store multiple value in a single variable
- It is an ordered collection of item.
- It is mutable (can be changed)
- It allows duplicate values
- It is hetrogenous (can store different type of value)
- In this indexing and alicing is applicable
- In both +ve and -ve indexing works
- It is created by using square brackets.

# Accessing Elements : 


```python
# Positive indexing
li= [1,1,1,2,2,"suhani",True,2.2]
print(li[0])
print(li[5])
```

```python
# Negative indexing : start from -1 and from end of a list
print(li[-1])
```

# Slicing :

- It is used to print the group of character of items from a list
- there are 3 parameters - start : stop : step

```python
li= [1,1,1,2,2,"suhanitext",True,2.2]
li[5:]
```

```python
li= [1,1,1,2,2,"suhanitext",True,2.2]
li[:6]
```

```python
li= [1,1,1,2,2,"suhanitext",True,2.2]
li[::-1]
```

```python
li= [1,1,1,2,2,"suhanitext",True,2.2]
li[::2]
```

# Traversing a list

```python
li= [1,1,1,2,2,"suhanitext",True,2.2]
for i in li:
    print (i)
```

```python
li= [1,1,1,2,2,"suhanitext",True,2.2]
for i in range (0,len(li)):
    print(li[i])
```

# Shallow Copy :

```python
a = [1,2,3]
b = a

b[0]=10
print(a)
print(b)
```

```python
# deep copy

import copy
a = [1,2,3]

b = copy.deepcopy(a)

b[0]=100

print(a)
print(b)
```

# List Comprehension in Python :

- It is a concise way to create a list using for loop or if statement in a sqaure bracket

```python
l = [1,2,3,4,5]
```

```python
l= []
for i in range(1,6):
    l.append(i)

l
```

```python
li = [i for i in range(1,6)]
li
```

```python
li = [i*i for i in range(1,6)]
li
```

```python
li = [i for i in range(1,21) if i%2==0] 
li
```

```python
num = [1,2,3,4,5,6,7,8]
res = ["Even" if  i %2==0 else "Odd" for i in num]
res
```
