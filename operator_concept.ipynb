# Logical Operator

```python
maths = 54
science = 45
english = 70
computer = 65
hindi = 35

res = maths > 40 and science >40 and english > 40 and hindi>40 and computer>40
print(res)
```

```python
maths = 4
science = 4
english = 7
computer = 65
hindi = 35

res = not(maths > 40 or science >40 or english > 40 or hindi>40 or computer>40)
print(res)
```

# Membership operator :

```python
l = [1,2,3,4,43,5,66,7,9,89,0]
print(66 not in l)
```

```python
name = ["Suhani","Ayushi","Amit","Sumit"]
print("Suhani" in name)
```

```python
my_dict = {"Name":"Suhani",
          "Age": 20
          }
res = "Name" in my_dict
print(res)
```

```python
my_dict = {"Name":"Suhani",
          "Age": 20
          }
res = "Suhani" in my_dict.values()
print(res)
```
