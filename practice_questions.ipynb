# Vsariables & Basic Maths

### Checking wheather the input provided by the user is even or odd.

```python
num = int(input("Enter the number:"))
if num % 2 ==0 :
    print (num,"is even.")
else :
    print (num,"is odd.")
```

### Temprature convertor from celsius to fehrenheit.

```python
c = float(input("Enter the temp in celsius:"))
temp = (c*(9/5)+32)
print(f'{temp} degree fehrenheit')
```
