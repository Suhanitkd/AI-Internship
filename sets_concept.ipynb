# Sets in Python

- A set is built-in data type + used to store multiple unique values in a single variable
- Does not allow duplicate value
- It is an unordered collection of item
- Does not support indexing and slicing 
- Is mutable (you can add or remove elements)
- In this data is stored in curly braces {}
- Can store different data types
- faster searching than list (by using hashing)
- Mutable element can't be stored in a set.

# Mathematical set operations:

```python
# union () : It is used to combine two sets.ipynb

s1 = {1,2,3,4}
s2 = {4,5,6,7}

print(s1.union(s2))
```

```python
# pipe operator :|
print(s1|s2)
```

```python
# Intersection : finds Common elements
s1 = {1,2,3,4}
s2 = {4,5,6,7}
print(s1.intersection(s2))
print(s1&s2)
```

```python
# difference () : return unique element of required set..
s1 = {1,2,3,4}
s2 = {4,5,6,7}

print(s1.difference(s2))
print(s2.difference(s1))
print(s1-s2)
```

```python
# symmetric difference : It return unique from both set
s1 = {1,2,3,4}
s2 = {4,5,6,7}
print(s1.symmetric_difference(s2))
print(s1^s2)
```

```python
# Frozenset : It is an immutable set. (we cannot add or remove element from a frozenset)

s= frozenset({1,2,3,4,5})
s
```
