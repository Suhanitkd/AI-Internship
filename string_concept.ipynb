# Strings :
A string is a sequence of character or a group of character enclosed withing single quotes ('') or double quotes ("") and triple quotes (''' ''' or """ """)

- A string can contain : 
 * alphabets
 * numbers
 * symbols 
 * spaces
 * special character

- String is Immutable(unchangeable)

 In String indexing and slicing is applicable :

 - Indexing :It is a position value of each character, in python there both +ve and -ve indexing
 - +ve indexing starts from 0 and goes upto n numbers, it starts from left side of a string
 - -ve indexing starts from -1 and from end of a string (right side of a string)

# Frequency of each character

```python
text = "Hello my self Anukul Srivastava"
for i in text:
    print (i,"=",text.count(i))
```

```python
text = "Hello my self Anukul Srivastava"
unique = set(text)
for i in unique:
    print(i,"= ",text.count(i))
```

# Write a program to find number of occurance of each unique character without using set function

```python
text = "Hello my self Anukul Srivastava"
var = " " # empty variable to store visited character
for i in text :
    if i not in var:
        print(i,"=",text.count(i))
        var +=i
var
```
