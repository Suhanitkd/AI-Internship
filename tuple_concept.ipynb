# Tuple : 

- Tuple is built -in data type in python
- It is an ordered collection of items
- It is hetrogenous
- In this we can store duplicate values.
- In this indexing and slicing is applicable
- It is created by using parentheses ()
- Tuples are immutable (un-changeable)
- Tuples works faster than list
- It is necessary for a tuple to have commas (otherwise it will behave like a noramal data-type)

```python
# Traverse tuple:

fruits = ("apple","pineapple","banana","strawberry","Mango","Cherry","Dark chocolate")
for i in fruits:
    print(i)
```

```python
fruits = ("apple","pineapple","banana","strawberry","Mango","Cherry","Dark chocolate")
for i in range (fruits):
    print (fruits[i])
```

```python
fruits = ("apple","pineapple","banana","strawberry","Mango","Cherry","Dark chocolate")
fruits[0] = "Grapes" #this shows that assignment is not possible in tuple
fruits
```

# Tuple unpacking

```python
student = ("Suhani","Electrical",20)
name,branch,age = student

print(name)
print(branch)
print(age)
```

```python
a = 5
b = 6

a,b = b,a
print(a,b)
```

```python
t1 = (1,2)
t2 = (3,4)
print(t1+t2)
```

```python
# Nested Tuple :

t = (1,2,(3,4))
print(t[2][1])
```
